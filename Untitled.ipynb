```python
# Fp1 - основная функция

def fp1(theta, gamma, epsilon1, epsilon11, L):
    part1 = (2 * ALPHA)/PI 
    part2 = (theta**3)/((gamma**-2)+theta**2)**2
    part3 = ((abs(1 - epsilon1 - i*epsilon11))**2)/((abs(theta**2 + gamma**-2 + 1 - epsilon1 - i*epsilon11))**2) 
    part4 = ((1 - e)**(-omega * (epsilon11/2) * L))**2
    part5 = (4 - e)**(-omega * (epsilon11/2) * L) 
    part6 = (math.sin(((gamma**-2) + (theta**2) + 1 - epsilon1)*((omega*L)/4)))**2
    return part1 * part2 * part3 * (part4 + part5 * part6)

```

```python
# func and get x/y func
def get_y(theta, gamma, x1):
    return ALPHA/(PI**2) * theta ** 2 * (1/(theta** 2 + 1/(gamma**2)) - 1/(theta**2 + 1/(gamma**2) - x1))**2


def get_XY_for_plot(theta_cnt, gamma, x1, min_theta=-0.02, max_theta=0.02):
    x = np.linspace(min_theta, max_theta, theta_cnt )

    y = np.copy(x)
    for idx, theta in enumerate(x):
        y[idx] = get_y(theta, gamma, x1)
    return x, y
```

```python
# graph
x_1, y_1 = get_XY_for_plot(1000, 300, X1)
x_2, y_2 = get_XY_for_plot(1000, 400, X1)


df = pd.DataFrame({
    "theta": np.concatenate([x_1, x_2]),
    "y": np.concatenate([y_1, y_2]),
    "label": ["График 1"] * len(x_1) + ["График 2"] * len(x_2)
})

fig = px.line(df, x="theta", y="y", color="label")


fig.update_layout(
    template="plotly_white"
)
fig.show()
```

```python
class Solution:
    def validMountainArray(self, arr: List[int]) -> bool:
        is_up = True
        if len(arr) < 3:
            return False
        if arr[0] > arr[1] or arr[0] == arr[1]:
                return False
        for i in range(1, len(arr) - 1):
            if arr[i] == arr[i + 1]:
                return False
            if is_up and arr[i] > arr[i + 1]:
                is_up = False
            elif (not is_up) and arr[i] < arr[i + 1]:
                return False
        return not is_up

```

```python
from math import inf

class Solution:
    def findDisappearedNumbers(self, nums: List[int]) -> List[int]:
        st = set(nums)
        arr = []
        for num in range(1, len(nums) + 1):
            if num not in st:
                arr.append(num)
        return arr.reverse()

```
