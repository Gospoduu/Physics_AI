```python

def variables(data, EeV):
    Z = data["Z"].iloc[0]
    N = data["N"].iloc[0]
    ro = data["ro"].iloc[0]
    chi = data["chi"].iloc[0]
    # print("Z =", Z, "N =", N, "ro =", ro,"chi =", chi)

    chi_string = chi.replace("'", '"')
    chi_dct = json.loads(chi_string)
    dict_values = [float(i) for i in chi_dct]
    dict_values[:5]

    m_e = 0.18 * (10**-24) 
    m_a = (N * m_p_n + Z * m_e)
    n0_m = (ro / (5 * m_a))
    k_n0_m = (n0_m * 10**4) / 125
    om2 = ((4 * PI * Z) / 137) * k_n0_m

    EeV_s = min_mse(EeV, dict_values)
    omega = EeV_s * 10**5 /2  # 
    omega0 = (om2**0.5) * 10**5 /2

    f1, f2 = chi_dct[str(EeV_s)].values()
    print(f"f1 = {f1}, f2 = {f2}")
    X1 = (-f1/Z )* (omega0/omega)**2

    xi_1  = (-1/Z)*(f1)*(omega0**2 / omega**2)  # Fp1

    xi_11 = (-1/Z)*(f2)*(omega0**2 / omega**2)  # Fp1
    print(f"xi_1 = {xi_1}, xi_11 = {xi_1} = {(-1/Z)} * {(f2)} * {omega0**2}/{omega**2}")
    print(f"Z = {Z}, N = {N}, ro = {ro}")

    epsilon1 = 1 + xi_1  # epsilon'
    epsilon11 = xi_11  # epsilon"
    epsilon = epsilon1 + 1.j * xi_11
    print(f"eps1 = {epsilon1}, eps11 = {epsilon11}")
    return epsilon1, epsilon11
```

```python
# Fp1 - основная функция
from numpy import e

def fp1(data, theta, gamma, L):
    epsilon1, epsilon11 = variables(data, EeV)
    epsilon = epsilon1 + 1.j * epsilon1
    part1 = (2 * ALPHA)/PI 
    part2 = (theta**3)/(((gamma**-2)+theta**2)**2)
    part3 = ((abs(1 - epsilon1 - 1.j*epsilon11))**2)/((abs(theta**2 + gamma**-2 + 1 - epsilon1 - 1.j*epsilon11))**2)
    part4 = ((1 - e)**((-omega * (epsilon11/2) * L))**2)
    part5 = (4 * e)**(-omega * (epsilon11/2) * L)
    part6 = (math.sin(((gamma**-2) + (theta**2) + 1 - epsilon1)*((omega*L)/4)))**2
    print(f"part1: {part1}, part2: {part2}, part3: {part3}, part4: {part4}, part5: {part5}, part6: {part6}")
    return part1 * part2 * part3 * (part4 + part5 * part6)

```

```python
def get_XY_for_fp1_plot(theta_cnt, data, gamma, L, min_theta=0, max_theta=0.1):
    x = np.linspace(min_theta, max_theta, theta_cnt )

    y = np.copy(x)
    for idx, theta in enumerate(x):
        y[idx] = fp1(data, theta, gamma, L)
    return x, y
```

```python
# func and get x/y func
def get_y(theta, gamma, x1):
    return ALPHA/(PI**2) * theta ** 2 * (1/(theta** 2 + 1/(gamma**2)) - 1/(theta**2 + 1/(gamma**2) - x1))**2


def get_XY_for_plot(theta_cnt, gamma, x1, min_theta=-0.02, max_theta=0.02):
    x = np.linspace(min_theta, max_theta, theta_cnt )

    y = np.copy(x)
    for idx, theta in enumerate(x):
        y[idx] = get_y(theta, gamma, x1)
    return x, y
```

```python
# graph
x_1, y_1 = get_XY_for_plot(1000, 300, X1)
x_2, y_2 = get_XY_for_plot(1000, 400, X1)


df = pd.DataFrame({
    "theta": np.concatenate([x_1, x_2]),
    "y": np.concatenate([y_1, y_2]),
    "label": ["График 1"] * len(x_1) + ["График 2"] * len(x_2)
})

fig = px.line(df, x="theta", y="y", color="label")


fig.update_layout(
    template="plotly_white"
)
fig.show()
```
