```python
import pandas as pd
import numpy as np
import os
import json

periodic_table = pd.read_csv(r"/path/Physics_AI/data/Periodic Table of Elements.csv")
periodic_table
```

```python
dir_path = r"/path/Physics_AI/data/chi"
dict_data = {}
for i in os.listdir(dir_path):
    with open(dir_path + "/" + i, 'r') as f:
        symb = i.split(".")[0]
        symb = symb[0].upper() + symb[1:].lower()
        data = json.load(f)
        dict_data[symb] = str(data)
        print(symb)
        

        
```

```python
dict_data
pd.DataFrame.from_dict(dict_data, orient='index').to_csv(r"/path/Physics_AI/data/chi.csv")
```
